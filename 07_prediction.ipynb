```python
import pandas as pd
import joblib
```

```python
model = joblib.load("models/final_model.pkl")
scaler = joblib.load("models/scaler.pkl")
```

```python
df = pd.read_csv("csv//processed_data.csv")

customer = df.drop("Churn", axis=1).iloc[[0]]
```

```python
customer
```

```python
customer_scaled = scaler.transform(customer)

prediction = model.predict(customer_scaled)

probability = model.predict_proba(customer_scaled)
```

```python
print("Prediction:", prediction[0])

print("Probability of No Churn :", probability[0][0])
print("Probability of Churn    :", probability[0][1])
```

```python
customers = df.drop("Churn", axis=1).head(10)

customers_scaled = scaler.transform(customers)

predictions = model.predict(customers_scaled)

customers["Prediction"] = predictions

customers
```

```python
customers["Churn Probability"] = model.predict_proba(customers_scaled)[:,1]

customers.head()
```
