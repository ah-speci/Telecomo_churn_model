```python
df=pd.read_csv('csv//tele_model_totalcharges_changed.csv')
```

```python
df['SeniorCitizen']=df['SeniorCitizen'].replace({1: 'Yes', 0: 'No'})    
```

```python
df['SeniorCitizen'].value_counts()
```

```python
senior_citizen_churn_rate=(pd.crosstab(df['SeniorCitizen'],df['Churn'],normalize='index')*100).round(2)
print(senior_citizen_churn_rate)
```

```python
senior_citizen_churn_rate.plot(kind='bar',stacked=True)
plt.ylabel('Percentage')
plt.title('Churn Rate by Senior Citizen Status')
plt.show()
```

```python
plt.scatter(df['tenure'],df['TotalCharges'],c=df['MonthlyCharges'])
plt.colorbar(label='Monthly Charges')
plt.xlabel('Tenure')
plt.ylabel('Total Charges')     
plt.title('Tenure vs Total Charges')   
plt.show() 
```

```python
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ["0-12", "13-24", "25-36", "37-48", "49-60", "61-72"]

df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
df[['tenure','TenureGroup']].head(10)
```

```python
churn_rate_by_tenure_group = (pd.crosstab(df['TenureGroup'], df['Churn'], normalize='index') * 100).round(2)
print(churn_rate_by_tenure_group)
```

```python
churn_rate_by_tenure_group.plot(kind='bar', stacked=True)
plt.ylabel('Percentage')
plt.title('Churn Rate by Tenure Group')
plt.grid(axis='y')
plt.show()
```

```python
df['TotalCharges'].hist(bins=30)
plt.xlabel('Total Charges')
plt.ylabel('Frequency')
```

```python
bins = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000,8000, 9000, 10000]
labels = ["0-1000", "1001-2000", "2001-3000", "3001-4000", "4001-5000", "5001-6000", "6001-7000","7001-8000","8001-9000","9001-10000"]  
df['TotalChargesGroup'] = pd.cut(
    df['TotalCharges'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
df[['TotalCharges','TotalChargesGroup']].head(10)
```

```python
churn_rate_by_total_charges_group = (pd.crosstab(df['TotalChargesGroup'], df['Churn'], normalize='index') * 100).round(2)   
print(churn_rate_by_total_charges_group )
```

```python
churn_rate_by_total_charges_group.plot(kind='bar',stacked=True)
plt.ylabel('Percentage')
plt.xlabel('Total Charges Group')
plt.title('Churn Rate by Total Charges Group')
plt.legend(title='Churn')
plt.show()
```
