```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('csv//day2_output.csv')
```

```python
plt.figure(figsize=(7,5))
plt.hist(df[df["Churn"]=="Yes"]["MonthlyCharges"], alpha=0.5, label="Yes")
plt.hist(df[df["Churn"]=="No"]["MonthlyCharges"], alpha=0.5, label="No")
plt.xlabel("Monthly Charges")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
internet_churn=pd.crosstab(df["InternetService"],df["Churn"],normalize="index")*100
internet_churn.plot(kind="bar", stacked=True, figsize=(7,5) )
```

```python
df['PaymentMethod'].value_counts()
payment_churn=pd.crosstab(df["PaymentMethod"],df["Churn"],normalize="index")*100
payment_churn.plot(kind="bar", stacked=True, figsize=(7,5) )
```

```python
def churn_rate(column1,column2):
    churn_rate=pd.crosstab(df[column1],df[column2],normalize="index")*100
    churn_rate.plot(kind="bar", stacked=True, figsize=(7,5) )
    plt.title(f"Churn Rate by {column1}")
    plt.xlabel(column1) 
    plt.ylabel("Percentage")    
```

```python
churn_rate("PaperlessBilling","Churn")
```

```python
churn_rate("PhoneService","Churn")
```

```python
def box(column1):
    plt.figure(figsize=(7,5))
    df.boxplot(column=column1)
    plt.title(f"Boxplot of {column1}")
    plt.suptitle("")
    plt.ylabel(column1)
    plt.show()
```

```python
box("MonthlyCharges")
box("TotalCharges")
box("tenure")
```

```python
encoded = df.copy()

for col in encoded.select_dtypes(include="object"):
    encoded[col] = encoded[col].astype("category").cat.codes

corr = encoded.corr(numeric_only=True)
```

```python
import seaborn as sns

plt.figure(figsize=(16,12))
sns.heatmap(corr,cmap="coolwarm")
plt.show()
```
